# PyCity Schools Analysis

**Summary**

This analysis considers data about math and reading scores from the 15 schools present in the district.

Across both types of schools, charter schools performed better than district schools when observing math, reading, and overall pass rates. This is because charter schools displayed common trends with the ones correlating with achieving higher pass rates. The trends are:
- School sizes below 2000 students experience a better math and reading academic performance.
- Spending budgets less $585 per student results in a better math and reading academic performance.  

Furthermore, across all schools in the district, the overall percentage of students passing both math and reading was weighed down by the percentage of students passing math. This passing rate was significantly lower than the percentage of students passing reading so more focus is required on math. 

- **Conclusion one:** 

Schools which had less than 2000 students experienced a significantly higher percentage of students passing both reading and math (approximately 31-32% higher than schools with 2000-5000 students). 

It was also found that spending more than $585 per student would not equate to a higher passing rate in either subject, in fact the higher the budget per student, the more progressively worse passing rates became; the bottom performing schools (based on the overall passing rate of both math and reading) all spent over \\$630 per student. 

⅞ of the charter schools analysed had less than 2000 students and spent less than \\$630 per student. Whereas 6/7 of the district schools did have over 2000 students and spent more than \\$630 per student. Therefore, it is recommended not to budget above \\$630 per student if you wish to achieve a higher passing rate in both math and reading within that school.

To depict how much of a difference this decision could make, Charter schools had a overall passing % (for both math and reading) of 90.43% whereas District schools had a overall passing % (for both math and reading) of only 53.67%.



- **Conclusion two:**

More focus is needed on increasing the passing rate of math per school.  The percentage of students passing reading across all schools in the district was 10.82% higher than the percentage of students passing math. 

When observing the average math score per grade across all schools in the district, it was found that grade 10 had the most occurrences of the minimum math score across al years. Therefore, it would be wise to focus on this grades performance in math.

Although this analysis is limited to being able to visualise the amount of budget allocated to math and reading resources, if it were found that there were to be a deficit in maths allocated amount, this is something which could be prioritised such as hiring more expeirenced staff and resources for math. 



```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
$585 per student would not equate to a higher passing rate in either subject; the bottom performing schools all spent over $630 per student. 
⅞ of the charter schools analysed had less than 2000 students and spent less than $630 per student. Whereas 6/7 of the district schools did have over 2000 students and spent more than $630 per student. Therefore, it is recommended not to budget above $630 per student if you wish to achieve a higher passing rate in both math and reading within that school. 

# Stores filepath in a variable
schools_complete_csv = Path("Resources/schools_complete.csv")
students_complete_csv = Path("Resources/students_complete.csv")

# Reads our schools and students data with the pandas library and stored each file into a DataFrame (no encoding required).
school_data = pd.read_csv(schools_complete_csv)
student_data = pd.read_csv(students_complete_csv)

# Merges both the school and students DataFrames
# Any school names present in school_data but not students_data won't be merged as we do not have data for them to analyse.    
schools_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
schools_complete_df.head()
```

# District Summary

```python
# Calculate the total number of unique schools
school_count =len(schools_complete_df["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = schools_complete_df["Student ID"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = schools_complete_df["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = schools_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = schools_complete_df[(schools_complete_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading 
passing_reading_count = schools_complete_df[(schools_complete_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = schools_complete_df[
    (schools_complete_df["math_score"] >= 70) & (schools_complete_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_df = pd.DataFrame({
                                    "Total Schools" : [school_count],
                                    "Total Students": [student_count],
                                    "Total Budget" : [total_budget],
                                    "Average Math Score" : [average_math_score], 
                                    "Average Reading Score" : [average_reading_score], 
                                    "% Passing Math" : [passing_math_percentage], 
                                    "% Passing Reading" : [passing_reading_percentage], 
                                    "% Overall Passing" : [overall_passing_rate]
                                    })

# Formatting
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary_df
```

```python
# Calculate the total student count per school
per_school_counts = school_data.rename(columns={"school_name":"School Name", "size":"Total Students"})
per_school_counts = per_school_counts[["School Name", "Total Students"]]
per_school_counts
```

```python
# Calculate the total student count per school
#per_school_counts = school_data.rename(columns={"school_name":"School Name", "size":"Total Students"})
#per_school_counts = per_school_counts[["School Name", "Total Students"]]



# Calculate the total school budget and per capita spending per school
#Total school budget
per_school_budget = school_data.rename(columns={"school_name":"School Name", "budget":"Total School Budget"})
per_school_budget = per_school_budget[["School Name", "Total School Budget"]]

#Per capita spending
per_school_capita = pd.merge(per_school_counts, per_school_budget, on=["School Name", "School Name"])
per_school_capita["Per Student Budget"] = per_school_capita["Total School Budget"]/ per_school_capita["Total Students"]
per_school_capita = per_school_capita[["School Name", "Per Student Budget"]]

#Combined
per_school_budget_capita = pd.merge(per_school_budget, per_school_capita, on=["School Name", "School Name"])
per_school_budget_capita.set_index("School Name")
per_school_budget_capita
```

```python
# Calculate the average test scores per school
#math
grouped_schools = student_data.groupby("school_name")
per_school_math = pd.DataFrame(grouped_schools["math_score"].mean())
per_school_math["School Name"] = per_school_math.index
per_school_math = per_school_math.reset_index(drop=True)

per_school_math = per_school_math[["School Name", "math_score"]] 
per_school_math = per_school_math.rename(columns={"math_score":"Average Math Score"})

per_school_math

#reading
per_school_reading = pd.DataFrame(grouped_schools["reading_score"].mean())
per_school_reading["School Name"] = per_school_reading.index
per_school_reading = per_school_reading.reset_index(drop=True)

per_school_reading = per_school_reading[["School Name", "reading_score"]] 
per_school_reading = per_school_reading.rename(columns={"reading_score":"Average Reading Score"})

per_school_reading

per_school_math_reading = pd.merge(per_school_math, per_school_reading, on="School Name")
per_school_math_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = student_data.loc[student_data["math_score"] >= 70, :]
grouped_school_names = students_passing_math.groupby("school_name")
school_students_passing_math = pd.DataFrame(grouped_school_names["Student ID"].count())
school_students_passing_math["School Name"] = school_students_passing_math.index
school_students_passing_math = school_students_passing_math.reset_index(drop=True)
school_students_passing_math = school_students_passing_math[["School Name", "Student ID"]]
school_students_passing_math = school_students_passing_math.rename(columns={"Student ID":"Total Students Passing Math"})
school_students_passing_math

```

```python
# Calculate the number of students per school with reading scores of 70 or higher

students_passing_reading =  student_data.loc[student_data["reading_score"] >= 70, :]
grouped_school_names = students_passing_reading.groupby("school_name")
school_students_passing_reading = pd.DataFrame(grouped_school_names["Student ID"].count())
school_students_passing_reading
school_students_passing_reading["School Name"] = school_students_passing_reading.index
school_students_passing_reading = school_students_passing_reading.reset_index(drop=True)
school_students_passing_reading = school_students_passing_reading[["School Name", "Student ID"]]
school_students_passing_reading = school_students_passing_reading.rename(columns={"Student ID":"Total Students Passing Reading"})
school_students_passing_reading
```

```python
# Calculate the number of students per school that passed both math and reading with scores of 70 or higher

students_passing_math_and_reading =  student_data.loc[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70), :]

grouped_school_names = students_passing_math_and_reading.groupby("school_name")
students_passing_math_and_reading = pd.DataFrame(grouped_school_names["Student ID"].count())
students_passing_math_and_reading
students_passing_math_and_reading["School Name"] = students_passing_math_and_reading.index
students_passing_math_and_reading = students_passing_math_and_reading.reset_index(drop=True)
students_passing_math_and_reading = students_passing_math_and_reading[["School Name", "Student ID"]]
students_passing_math_and_reading = students_passing_math_and_reading.rename(columns={"Student ID":"Total Students Passing Math and Reading"})
students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
# % Passing Math

passing_math_percentage = pd.merge(per_school_counts, school_students_passing_math, on="School Name")
passing_math_percentage["% Passing Math"] = (passing_math_percentage["Total Students Passing Math"] /  passing_math_percentage["Total Students"]*100)
passing_math_percentage = passing_math_percentage[["School Name", "% Passing Math"]]
passing_math_percentage


# % Passing Reading

passing_reading_percentage = pd.merge(per_school_counts, school_students_passing_reading, on="School Name")
passing_reading_percentage["% Passing Reading"] = (passing_reading_percentage["Total Students Passing Reading"] /  passing_reading_percentage["Total Students"]*100)
passing_reading_percentage = passing_reading_percentage[["School Name", "% Passing Reading"]]
passing_reading_percentage

# % Passing Overall

passing_math_reading_percentage = pd.merge(per_school_counts, students_passing_math_and_reading, on="School Name")
passing_math_reading_percentage["% Overall Passing"] = (passing_math_reading_percentage["Total Students Passing Math and Reading"] / passing_math_reading_percentage["Total Students"]*100)
passing_math_reading_percentage = passing_math_reading_percentage[["School Name", "% Overall Passing"]]
passing_math_reading_percentage

passing_math_reading_percentages = pd.merge(passing_math_percentage, passing_reading_percentage, on="School Name")
passing_percentages = pd.merge(passing_math_reading_percentages, passing_math_reading_percentage, on="School Name")
passing_percentages
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df.head()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
del school_spending_df["Per Student Budget"]
school_spending_df = pd.merge(school_spending_df, per_school_capita, on="School Name") 

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels, include_lowest=True)

school_spending_df = school_spending_df.set_index("School Name")
school_spending_df = school_spending_df[["School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing", "Spending Ranges (Per Student)"]]
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].map("${:,.2f}".format)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean())
spending_reading_scores = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean())
spending_passing_math = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean())
spending_passing_reading = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean())
overall_passing_spending = pd.DataFrame(school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean())
```

```python
# Assemble into DataFrame
Merge_1 = pd.merge(spending_math_scores, spending_reading_scores, on="Spending Ranges (Per Student)")
Merge_2 = pd.merge(Merge_1, spending_passing_math, on="Spending Ranges (Per Student)")
Merge_3 = pd.merge(Merge_2, spending_passing_reading, on="Spending Ranges (Per Student)")

spending_summary = pd.merge(Merge_3, overall_passing_spending, on="Spending Ranges (Per Student)")
spending_summary
```
